```python
#Importing Packages
import keras
from keras.models import Sequential
from keras.layers import Dense, Flatten
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
#Importing Dataset
from keras.datasets import mnist
(train_img, train_lab), (test_img, test_lab) = mnist.load_data()
```

```python
#Building Model
model=Sequential()
model.add(Flatten(input_shape=(28,28)))
model.add(Dense(300, activation='relu'))
model.add(Dense(200, activation='relu'))
model.add(Dense(100, activation='relu'))
model.add(Dense(10, activation="softmax"))
```

```python
#Compiling Model
model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=['accuracy'])
```

```python
#Fitting the Model
model.fit(train_img, train_lab, epochs=10)
model.save('model.h5')
```

```python
#Evaluate the Model
print(model.evaluate(test_img, test_lab))
```

```python
#Predicting First 10 test images
pred = model.predict(test_img[:10])
# print(pred)
p=np.argmax(pred, axis=1)
print(p)
print(test_lab[:10])
```

```python
#Visualizing prediction
for i in range(10):
  plt.imshow(test_img[i], cmap='binary')
  plt.title("Original: {}, Predicted: {}".format(test_lab[i], p[i]))
  plt.axis("Off")
  plt.figure()
```
